# What is Haberman Dataset

The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.
(https://www.kaggle.com/gilsousa/habermans-survival-data-set)

# Objective of the Analysis :

To classify/predict a patient survival who had undergone surgery for breast cancer.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the haberman dataset into a pandas dataframe. The CSV should be present in the same 
# folder as the notebook.

dataset = pd.read_csv('haberman.csv')
```

```python
# Number of data points available for analysis.

print(dataset.shape)
```

```python
# Column Names in the dataset.

print(dataset.columns)
```

Column Information 
* Age - Age of patient at time of operation (numerical).
* Year - Patient's year of operation (year - 1900, numerical).
* nodes - Number of positive axillary nodes detected (numerical).The axillary nodes are a group of lymph nodes located in the axillary (or armpit) region of the body.
* status - Survival status (class attribute) 1 = the patient survived 5 years or longer 2 = the patient died within 5 year.

We have three independent variables (Age, Year, Nodes) and one dependent variable (status).

```python
# How many data points for each class are present ?
dataset['status'].value_counts()
```

This is an unbalaced dataset as 75% of the data points belong to one class.

```python
dataset.head()
```

```python
# Dataset Summary

dataset.info()
```

```python
# Get a statistical summary of all the numerical columns in the dataset.
dataset.describe()
```

From the above summary we can see that :
* The people in the dataset were from ages 30-83 and 75% of the people were below 61 years.
* The operations occurred from years 58-69 and the operations see to be evenly split between years.
* The number of nodes range from 0 to 52. 52 seems like an outlier since 75% of people had upto 4 nodes.

# Univariate Analysis (1-d scatter plot, pdf, cdf, boxplot and violin plots)

Univariate analysis is the simplest form of analyzing data. “Uni” means “one”, "variate" means "variable or numeric variable" so, in other words your data has only one variable. It doesn't deal with causes or relationships (unlike regression) and it's major purpose is to describe; it takes data, summarizes that data and finds patterns in the data.

```python
# Splitting the dataset depending on the class variable.

dataset_status_1 = dataset.loc[dataset["status"] == 1]
dataset_status_2 = dataset.loc[dataset["status"] == 2]
```

* 1-D Scatter Plot by Age

```python
plt.plot(dataset_status_1["age"], np.zeros_like(dataset_status_1["age"]), 'o')
plt.plot(dataset_status_2["age"], np.zeros_like(dataset_status_2["age"]), '*')
plt.show()
```

* 1-D Scatter Plot by year.

```python
plt.plot(dataset_status_1["year"], np.zeros_like(dataset_status_1["year"]), 'o')
plt.plot(dataset_status_2["year"], np.zeros_like(dataset_status_2["year"]), '*')
plt.show()
```

* 1-D Scatter Plot by nodes

```python
plt.plot(dataset_status_1["nodes"], np.zeros_like(dataset_status_1["nodes"]), 'o')
plt.plot(dataset_status_2["nodes"], np.zeros_like(dataset_status_2["nodes"]), '*')
plt.show()
```

### PDF( Probability Density Function )'s.

```python
sns.FacetGrid(dataset, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();

plt.title("Histogram of age")
plt.ylabel("Density")

plt.show();
```

```python
sns.FacetGrid(dataset, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();

plt.title("Histogram of year")
plt.ylabel("Density")

plt.show();
```

```python
sns.FacetGrid(dataset, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();

plt.title("Histogram of nodes")
plt.ylabel("Density")

plt.show();
```

#### Observations :

* Both the 1-D scatter plots and the PDF's show that all the features are overlapping and no definitive conclusions can be made using just one.
* The feature that looks promising is nodes and the feature that we can probably look less at is the year.

## CDF (Cumulative density function's)

```python
# Plot of CDF by age.

labels = ["pdf of status-1", "cdf of status-2", "pdf of status-2", "cdf of status-2"]

counts, bin_edges = np.histogram(dataset_status_1['age'], bins=10, density=True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

plt.title("PDF and CDF for age.")
plt.xlabel("age")
plt.ylabel("% of person's")

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(dataset_status_2['age'], bins=10, density=True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(labels)

plt.show()
```

```python
# Plot of CDF by year.

labels = ["pdf of status-1", "cdf of status-2", "pdf of status-2", "cdf of status-2"]

counts, bin_edges = np.histogram(dataset_status_1['year'], bins=10, density=True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

plt.title("PDF and CDF for year.")
plt.xlabel("year")
plt.ylabel("% of person's")

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(dataset_status_2['year'], bins=10, density=True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(labels)

plt.show()
```

```python
# Plot of CDF by nodes.

labels = ["pdf of status-1", "cdf of status-2", "pdf of status-2", "cdf of status-2"]

counts, bin_edges = np.histogram(dataset_status_1['nodes'], bins=20, density=True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

plt.title("PDF and CDF for nodes.")
plt.xlabel("nodes")
plt.ylabel("% of person's")

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(dataset_status_2['nodes'], bins=20, density=True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(labels)

plt.show()
```

#### Observations:

* All the people who has more than 45 node have not survived.
* All the people above age 75 have not survived.
* The year doesn't seem to influence the class variable. 

## Box Plot

Box-plot with whiskers: another method of visualizing the  1-D scatter plot more intuitivey.
Box-plot gives you the statistical summery of data.
Rectangle represent the 2nd and 3rd quartile (horizontal line either side of the rectangle).
The horizontal line inside box represents median.

```python
sns.boxplot(x="status", y="age", hue="status", data=dataset)

plt.title("Box plot for survival status and age.")

plt.show()
```

```python
sns.boxplot(x="status", y="year", hue="status", data=dataset)

plt.title("Box plot for survival status and year.")

plt.show()
```

```python
sns.boxplot(x="status", y="nodes", hue="status", data=dataset)

plt.title("Box plot for survival status and nodes.")

plt.show()
```

## Violin Plot

A violin plot combines the benefits of the previous two plots and simplifies them.
Denser regions of the data are fatter, and sparser ones thinner in a violin plot.

```python
sns.violinplot(x="status", y="age", hue="status", data=dataset)
plt.title("Violin plot for survival status and age.")
plt.show()
```

```python
sns.violinplot(x="status", y="year", hue="status", data=dataset)
plt.title("Violin plot for survival status and year.")
plt.show()
```

```python
sns.violinplot(x="status", y="nodes", hue="status", data=dataset)
plt.title("Violin plot for survival status and nodes.")
plt.show()
```

### Observations:

*  From the box-plot we can see that all the people with age less than 35 survived.

# Bivariate Analysis (2-D scatter plot, pair plot)

Bivariate analysis is the simultaneous analysis of two variables (attributes). It explores the concept of relationship between two variables, whether there exists an association and the strength of this association, or whether there are differences between two variables and the significance of these differences.

## Scatter Plot

A scatter plot is a two-dimensional data visualization that uses dots to represent the values obtained for two different variables - one plotted along the x-axis and the other plotted along the y-axis.

```python
sns.set_style("whitegrid");
sns.FacetGrid(dataset, hue="status", size=5) \
   .map(plt.scatter, "age", "nodes") \
   .add_legend();

plt.title("2-D scatter plot of age and nodes.");
plt.show();
```

## Pair Plots.

Pair plot is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters. It also helps to form some simple classification models by drawing some simple lines or make linear separation in our dataset.

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(dataset, hue="status", vars = ["age", "year", "nodes"], size=4);
plt.suptitle("pair plot of age, year and nodes")
plt.show()
```

## Conclusions:

*  The given dataset is not lineraly separable from each class. one of the reasons may be because the dataset is not balanced i.e. it does not contain equal data points for each class.
* There are too much overlapping in the data-points and hence can't be separated by a straight line.
* We cannot build a simple model to determine the survival rate.
* The only feature that was promising was nodes.
